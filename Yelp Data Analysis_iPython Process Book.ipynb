### 1) cleaning check-in sets

```python
checkin = pd.read_csv("original_data/checkin.csv")
checkin.head()
```

```python
checkin.ix[:,0].values
```

```python
Days = []
for day in range(7):
    Day = np.zeros(len((checkin.ix[:,0]).values))
    for hour in range(24): 
        Day = Day + (checkin.ix[:, 2+hour+day*24]).values
    Days.append(Day)
cleaned_checkin = pd.DataFrame()
cleaned_checkin['business_id'] = checkin['business_id'].values

Hours = []
for hour in range(24):
    Hour = np.zeros(len((checkin.ix[:,0]).values))
    for day in range(7): 
        Hour = Hour + (checkin.ix[:, 2+hour+day*24]).values
    Hours.append(Hour)

cleaned_checkin['Weekdays'] = sum(Days[1:6])
cleaned_checkin['Weekends'] = sum([Days[0],Days[6]])
cleaned_checkin['Morning'] = sum(Hours[8:12])
cleaned_checkin['Afternoon'] = sum(Hours[12:17])
cleaned_checkin['Evening'] = sum(Hours[17:24])
cleaned_checkin['Midnight'] = sum(Hours[0:8])

for day in range(7):
    cleaned_checkin['Day%d' % day] = Days[day]

for hour in range(24):
    cleaned_checkin['Hour%02d' % hour] = Hours[hour]

cleaned_checkin.head()
```

```python
cleaned_checkin.to_csv("cleaned_data/cleaned_checkin.csv")
```

### 2) cleaning business hours

```python
attr = pd.read_csv("original_data//business_hours.csv")
attr.head()
```

```python
cols = list(attr.columns)
del cols[0]
del cols[0]
cols
```

```python
# Get most frequent hours for all possible open/close days. I will fill with this, if time is not given. 
# Assumption here is that best guess is that business opens/closes just as most of other business do.
typical = {}

for k in cols:    
    cleanedList = [x for i,x in enumerate(attr[k]) if not(attr[k].isnull()[i])]
    hours = [datetime.strptime(x, '%H:%M').hour for x in cleanedList]
    typical[k] = int(mode(hours)[0])
```

```python
for x in cols:
    print x, typical[x]
# Interesting results. Open on Thursdays and Tuesdays is most often at 11, while Monday, Wednesday, Friday are at 9. 
```

```python
for k in cols:
    if typical[k]<10:
        typical[k] = '0'+str(typical[k])+':00'
    else:
        typical[k] = str(typical[k])+':00'
for x in cols:
    print x, typical[x]
```

```python
# filling with typical times where all times are missing  #attr.index:

attr['missing times'] = [0]*len(attr.index)

for x in attr.index:   
    logic=False
    for c in cols:   
        logic = (not(str(attr.loc[x,c])=='nan')) | (logic)
        if logic:
            break
    
    if not(logic):
        for y in cols:
            attr.loc[x,y]=typical[y]
            attr.loc[x,'missing times']=1 # keep a binary variable reminding that this restourant had missing hours. 
                                          # Might be useful in predictions
```

```python
# filling in with False, where only some times is missing (because this probably means they do not work)
not_working_lists = {}
for c in cols:
    not_working_lists[c]=[]
    
for x in attr.index:
    for c in cols:
        if (str(attr.loc[x,c])=='nan') | (attr.loc[x,c]==False):
            attr.loc[x,c]=False
            not_working_lists[c].append(attr.loc[x,'business_id'])
```

```python
attr.to_csv('cleaned_data/cleaned_business_hours.csv')
attr[['business_id','missing times']].to_csv('cleaned_data/missing_times.csv')
```

```python
for i in range(0,len(city)):
    a=sub[sub.city==city[i]]
    total=a.shape[0]
    a=a.reset_index(drop=True)
    a["competitor"]=0
    a["competitor_index"]=0
    la_min=min(a.latitude)
    la_max=max(a.latitude)
    long_min=min(a.longitude)
    long_max=max(a.longitude)
    la_delta=(la_max-la_min)/10
    long_delta=(long_max-long_min)/10
    if (la_delta==0) or (long_delta==0):
        continue
    b=[[[],[],[],[],[],[],[],[],[],[]] for j in range(10)]
    for j in range(0,total):
        x=int((a.latitude[j]-la_min)/la_delta)
        if x==10:
            x=9
        y=int((a.longitude[j]-long_min)/long_delta)
        if y==10:
            y=9
        b[x][y].append(a["business_id"][j])
    for j in range(0,total):
        x=int((a.latitude[j]-la_min)/la_delta)
        if x==10:
            x=9
        y=int((a.longitude[j]-long_min)/long_delta)
        if y==10:
            y=9
        a["competitor"][j]=len(b[x][y])
    a["competitor_index"]=a["competitor"]/total
    for j in range(0,total):
        ind=a["Unnamed: 0"][j]
        sub["competitor"][ind]=a["competitor"][j]
        sub["competitor_index"][ind]=a["competitor_index"][j]
```

### 1) Random Forest Classifier

We decided to try fitting the data using Random Forest Classifier because it can handle correlated random variables, and the number of stars can be viewed as a categorical variable.

```python
from sklearn.cross_validation import cross_val_score

from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier as rfc

import seaborn as sns

data = pd.read_csv("cleaned_data/clean_combined.csv")
data.head()
```

#### Search for optimal criterion and optimal number of trees

```python
def predict_stars(n,features = range(11,108),method='gini'):
    X = data.ix[:,features]
    Y = pd.factorize(data.stars)[0]
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
    clf = rfc(random_state=0,max_features = 'log2',n_estimators=n,criterion=method)
    clf.fit(X_train,y_train)
    scores = cross_val_score(clf,X_test,y_test,cv=10)
    return scores

n_estimators = range(10,260,10)

# criterion = "Gini", different n_estimators

stars_gini = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = predict_stars(n_estimators[i])
    stars_gini[i,]=scores

# criterion = “entropy”, different n_estimators

stars_entropy = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = predict_stars(n=n_estimators[i],method='entropy')
    stars_entropy[i,]=scores

np.save("cleaned_data/stars_gini.csv",stars_gini)
np.save("cleaned_data/stars_entropy.csv",stars_entropy)
```

```python
sns.set_context("talk")
sns.set_style("white")

sns.boxplot(np.transpose(stars_gini))
plt.xlabel("Number of trees")
plt.ylabel("Score")
plt.title("Predicting Number of Stars using Random Forest Classifer (Gini)")
plt.xticks(range(1,len(n_estimators)+1),n_estimators)
plt.ylim(0.24,0.4)
```

```python
def predict_success(n,features = range(11,108),method='gini'):
    X = data.ix[:,features]
    Y = data.success
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
    clf = rfc(random_state=0,max_features = 'log2',n_estimators=n,criterion=method)
    clf.fit(X_train,y_train)
    return cross_val_score(clf,X,Y,cv=10)

# criterion = "Gini", different n_estimators

success_gini = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = predict_success(n_estimators[i])
    success_gini[i,]=scores
    
    
# criterion = “entropy”, different n_estimators

success_entropy = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = predict_success(n=n_estimators[i],method='entropy')
    success_entropy[i,]=scores
    
np.save("cleaned_data/success_gini.csv",success_gini)
np.save("cleaned_data/success_entropy.csv",success_entropy)
```

#### Search for optimal cutting value

```python
def cut(alpha,features = range(11,108)):
    X = data.ix[:,features]
    Y = data.success

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)

    clf = rfc(random_state=0,max_features = 'log2',n_estimators=180,criterion="gini")

    clf.fit(X_train,y_train)
    probs = clf.predict_proba(X_test)[:,1]
    y_pred = probs>alpha
    accuracy = np.mean(y_pred==y_test) 
    return accuracy

scores_cut = []

for i in arange(0.1,1.1,0.1):
    scores_cut.append(cut(i))

np.save("cleaned_data/scores_cut.csv",scores_cut)
    
fig, ax = plt.subplots()
ppl.scatter(ax,arange(0.1,1.1,0.1),scores_cut)

ax.legend()
ax.set_xlabel("Cutting value")
ax.set_ylabel("Score")
ax.set_title("Cutting value v.s. Random Forest Classifier Scores")
```

```python
# Function to get first look at model - how well it explains the data
def fit_model(model, data): 
    
    
    model.fit(data['train x'], data['train y'])
    y_pred = model.predict(data['test x'])
    correct_list = (y_pred == list(data['test y']))
    correct_list = [int(x) for x in correct_list]
    
    correct_proportion = round(sum(correct_list)/float(len(data['test y'])),4)
    total_correct = sum(correct_list)
    
    cv_score = cross_val_score(model, data['train x'], np.array(data['train y']), cv=10)
    
    metrics = [correct_proportion, total_correct, cv_score]
    metrics = [correct_proportion, total_correct, cv_score]
    
    return metrics 
```

```python
# Function to represent cross validation scores for differently parametrized models
def box_plot(data, names, title, xlab, ylab):
    
    avg_Acc = [0]*(len(data))
    for i in range(len(data)):
        avg_Acc[i] = np.mean(data[i])
    
    #palette = sns.color_palette("Blues", n_colors=len(names))
    #ranks = [avg_Acc.index(x) for x in sorted(avg_Acc)]
    #my_colors = [palette[i] for i in ranks]
    
    plt.figure(figsize=(12, 9))
    sns.set_style("white")  
    sns.set_style("ticks")
    sns.boxplot(data, names=names, color =  get_colorlist('blue', avg_Acc))
    sns.despine(top=True, right=True)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    plt.plot([x+1 for x in range(len(data))], avg_Acc, label= 'Average Accuracy score')
    plt.legend(fontsize=14)
    plt.title(title, fontsize=18, fontweight='bold')
    plt.xlabel(xlab, fontsize=16)
    plt.ylabel(ylab, fontsize=16)
    plt.show()
```

```python
# Does a summary of model accuracy for specified range of parameters. Return boxplot.
def do_summary(model, params, data):
           
    scores = [0]*len(params)
    
    for i,x in enumerate(params):
        if (type(model)==sklearn.svm.classes.SVR) | (type(model)== sklearn.svm.classes.SVC) | (type(model) == sklearn.linear_model.logistic.LogisticRegression):
            scores[i] = fit_model(model.set_params(C=x), data)[2]
            name = 'C'
        elif (type(model)== sklearn.tree.tree.DecisionTreeClassifier):
            scores[i] = fit_model(model.set_params(max_depth=x), data)[2]
            name = 'depth of tree'
        elif (type(model)== sklearn.ensemble.forest.RandomForestClassifier):
            scores[i] == fit_model(model.set_params(n_estimators=x),data)[2]
            name = 'trees'        
        elif (type(model)== sklearn.neighbors.classification.KNeighborsClassifier):
            scores[i] = fit_model(model.set_params(n_neighbors=x), data)[2]
            name = 'Number of neighbours'
        elif (type(model)==sklearn.linear_model.ridge.Ridge) | (type(model)==sklearn.linear_model.coordinate_descent.Lasso):
            scores[i] = fit_model(model.set_params(alpha=x), data)[2]
            name = 'alpha'
        else:
            return 'Unknown model'
        
    box_plot(scores, params, 'Searching for optimal '+ name, name, 'Accuracy score')
```

```python
# Function to get frequencies of stars in a vector. Important to have stars specified unifrmly across all testings (as integers)
def get_frequencies(vector, stars, relative=False):
    answer = {}    
    for i in stars:        
        answer[i] = sum([1 for x in vector if x==i])
    if relative:
        total = sum([answer[x] for x in answer.keys()])
        for i in stars:
            answer[i] = answer[i]/float(total)
    return answer
```

```python
# Splitting data into 6000 and 3000. This is for trying prediction where one model relies on output of another model
# (on second thought, could have used built in splitting function...)
def split_data(X, Y):
    
    full_data = X.reset_index(drop=True)
    full_y = pd.DataFrame(Y)
    
    reindexing = np.random.permutation(full_data.index)
    full_data = full_data.reindex(reindexing)
    full_y = full_y.reindex(reindexing)

    train_data = full_data.iloc[:6000,]
    train_data = train_data.reset_index(drop = True)
    test_data = full_data.iloc[6000:,]
    test_data = test_data.reset_index(drop = True)
    train_y = full_y.iloc[:6000,]
    train_y = train_y.reset_index(drop = True)
    test_y = full_y.iloc[6000:,]
    test_y = test_y.reset_index(drop = True)
    
    return train_data, test_data, list(train_y[0]), list(test_y[0])
```

```python
# Function to investigate accuracy and precision
def get_proportions(model, data, cv, regression):
    
    if cv:    
        # Do manual cross validation. Fit model on 75% of train data set, and then test it on remaining 25% of train data
        train_data, test_data, train_y, test_y = split_data(data['train x'], data['train y'])
        model.fit(train_data, train_y)
    
    else:
        test_data = data['test x']
        test_y = data['test y']
        test_y = pd.DataFrame(test_y)    
    
    y_pred = model.predict(test_data)
    
    if regression:  # condition when we use regression type models and not classifiers, 
                                    # which return floats. Need to convert back to categories
        y_pred = [int(round(x,0)) for x in y_pred]
        if (max(y_pred)>10) | (min(y_pred)<2):
            for i in range(len(y_pred)):
                if y_pred[i]>10:
                    y_pred[i] = 10
                elif y_pred[i]<2:
                    y_pred[i]=2
        
    correct_list = [int(x==y) for x,y in zip(y_pred,test_y)]

    df_connected = pd.DataFrame(test_y, list(correct_list))
    df_connected.reset_index(inplace=True)    
    df_connected.columns = ['correct', 'real']
    df_connected['predicted'] = list(y_pred)
    
    df_correct=df_connected[df_connected['correct']==1]
    
    star_freq = get_frequencies(test_y, range(2,11))         
    total_guesses = get_frequencies(y_pred, range(2,11))
    
    correct_guesses = {}
        
    for star in range(2,11):
        
        correct_guesses[star] = len(df_correct[df_correct['real']==star])         
              
    recall = [0]*9
    precision = [0]*9
    recall2 = [0]*9
    number_of_guesses = [0]*9
    
    for j,i in enumerate(range(2,11)):
        
        if star_freq[i]!=0:            
            recall[j] = round(correct_guesses[i]/float(star_freq[i]),3)
            recall2[j] = correct_guesses[i]
        else:
            recall[j] = 0
            recall2[j] = 0
            
        if total_guesses[i]!=0:            
            precision[j] = round(correct_guesses[i]/float(total_guesses[i]),3)
            number_of_guesses[j] = total_guesses[i]
        else:
            precision[j] = 0
            number_of_guesses[j] = 0
            
    return recall, precision, recall2, number_of_guesses
```

```python
# Iterating get_proportions function to get cross validation
def get_proportions_cv(model, data, cv, regression):
    
    recall = [0]*len(stars)
    precision = [0]*len(stars) 
    recall2 = [0]*len(stars) 
    number_of_guesses = [0]*len(stars)
    
    final_answer = [recall, precision, recall2, number_of_guesses]
    
    for i in range(cv):
        single_result = get_proportions(model, data, True, regression)
        for j in range(4):
            final_answer[j] = [x+y for x,y in zip(final_answer[j], single_result[j])]
    
    for i in range  (4):  
        final_answer[i] = [x/float(cv) for x in final_answer[i]]
    
    return final_answer
```

```python
# Fit list of models to chosen dataset
def fit_models(models, data):
    
    fitted_models = {}
    
    for key in models.keys():
        
        if type(models[key])!=list:
            fitted = models[key].fit(data['train x'], data['train y'])
            fitted_models[key] = fitted
            
        else:
            fitted_models[key] = []
            for i in range(len(models[key])):                
                fitted = models[key][i].fit(data['train x'], data['train y'])
                fitted_models[key].append(fitted)    
        
    return fitted_models
```

```python
# Get back subset of data frame which contains only specified stars, as well as list of stars. This is useful for plotting let's say just 
# worst and best restourants, and excluding the noise of mid-valued restourants. Or if you want to fit model to just categorise subset of
# stars, and not all of them at once.
def get_only(sub_stars, data, y_list):
    
    data = data[list(pd.DataFrame(y_list).isin(sub_stars)[0])]
    
    y_list = [x for x in y_list if x in sub_stars]
    
    return data, y_list
```

```python
# Normalizes data frame. Either to mean 0 and std. dev 1, or fit data into range [0,1] (if zero_one = True)
def normalize(data, zero_one = False):
    if zero_one:
         for column in data.columns:
                minimum = np.min(data[column])
                maximum = np.max(data[column])
                new_column = [(x-minimum) / (maximum - minimum) for x in data[column]]
                data[column] = new_column
         return data
    else:
        
        for column in data.columns:
            mean = np.mean(data[column])
            std = np.std(data[column])
            if std>0.00001:
                new_column = [(x-mean)/std for x in data[column]]
            else:
                new_column = data[column]
            data[column]=new_column
            if (max(data[column])>100000) | (min(data[column])<-100000):
                print column
        return data
```

```python
#Returns data frame with probabilities for each category
def get_probs(models, data, classes = False, use_key=True, log=False):

    if classes:
        cols = [-1,1]
    else:
        cols = [2,3,4,5,6,7,8,9,10]
        
    df_probs = pd.DataFrame(range(len(data.iloc[:,1])))

    for key in models.keys():
        if type(models[key])!=list:
            if use_key:
                if log:
                    model_probs = pd.DataFrame(models[key].predict_log_proba(data.values), 
                                           columns= [str(key)+str(x) for x in cols])
                else:
                    model_probs = pd.DataFrame(models[key].predict_proba(data.values), 
                                           columns= [str(key)+str(x) for x in cols])                
            else:
                if log:
                    model_probs = pd.DataFrame(models[key].predict_log_proba(data.values), 
                                           columns= [str(x) for x in cols])
                else:
                    model_probs = pd.DataFrame(models[key].predict_proba(data.values), 
                                           columns= [str(x) for x in cols])            
            df_probs = pd.concat([df_probs, model_probs], axis=1)            

        else:
            for i in range(len(models[key])):
                if log:
                    model_probs = pd.DataFrame(models[key][i].predict_log_proba(data.values), 
                                           columns= [str(key)+str(i)+str(x) for x in cols])                    
                else:
                    model_probs = pd.DataFrame(models[key][i].predict_proba(data.values), 
                                           columns= [str(key)+str(i)+str(x) for x in cols])
                df_probs = pd.concat([df_probs, model_probs], axis=1)            

    df_probs.drop(0, axis=1, inplace=True)
    
    return df_probs
```

```python
# Convert time to integer, as some models cannot handle input of type "11:00" etc.
def convert_time(row):
    for day in ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']:
        for action in ['.open','.close']:
            if row[day+action]=='False':
                row[day+action]=-1
            else:
                row[day+action] = datetime.strptime(row[day+action], '%H:%M').hour
    return row.tolist()

df = pd.DataFrame(columns=df_old.columns)

for i in range(len(df_old)):
    df.loc[i] = convert_time(df_old.ix[i,])
    
del df['missing times_x']
del df['Unnamed: 0.1']
    
df.to_csv("cleaned_data/convert_time.csv")

df.head()
```

#### Force predicted density to be as expected

One potential trick can be to force predicted relative frequency of stars to be as observed in training data set. For example if we know
that there are roughly 5% of starts = 1, 10% of stars = 2, etc., we can try to produce joint guesses so that they would adhere to these
proportions. This means that some times we will be guessing not according to maximum probability accross categories (if that category
already reached its relative frequency limit, we should not guess it again). It is not clear how to handle this. My idea is that I want
to maximise total sum of probabilities (each category you guess has associated probability with it) with constraint of final joint 
guesses' relative frequencies being equal to the relative frequencies observed in training data set. 

I am implementing this by taking max and second largest probability from each column, then sorting according to difference (difference is
like opportunity cost, because once you run out of category, you will have to guess another one, so if difference between max and
second largest probability is big, you want to get maximum, but if it is not big, you can guess second largest category without loosing
too much on total sum of probabilities).

The reason why this can work, is because we are using additional information. When model makes predictions, it treats each guess 
independently. But we roughly know what proportions are attributed to each star. So when simultaneously predicting many stars, maybe we
can use this knowledge to make more informative decisions.

## V. Reviews and Sentiment Analysis

** Reviews, or opinions of users, eveluate a restaurant from diverse aspects and may provide users with a holistic view besides certain attributes Yelp has included as attributes. In order to explore the information contained in review, we need a different approach to deal with the comparatively large data. The most typical difficulty to deal with reviews is that, they are strings of characters, and not naturally quantitative. Therefore we need to interpret strings by turning them into numbers appropriately. **

It is a natural to start our analysis or strings by breaking down the sentences into words. Since we notice reviews are mostly commentary or sentimental remarks, we expect to see both positive and negative adjectives or adverbs play a leading role. At the same time, we also expect to see some nouns or verbs with certain emotional hints.
<p>
However, those word may be overwhelmed by neutral words whose occurance is much prevalent. We here try to assign a score to each word, and use the score to distinguish meaningful words from meaningless ones. The implementation is to calculate the average rating of reviews in which a certian word appears. All words are ranked from the highest score to the lowest score, and then divided into positive, negative and neutral groups according to their score.
<p>
So far, each review is recognized as a collection of positive words, negative words and neutral words, and is associated with a rating given by the user. We can now apply machine learning toolkits, such as regression or classification on the data, to capture the correlation between review words and the ratings, and then lead to expected inference and prediction. 
<p>
Finally, we notice a piece of review is a random sample from all opinions towards that restaurant, we can expect the review as an unbiased estimator. Indeed, more reviews will lead to better precision in prediction, thus we collect all remarks towards a restaurant and use the averaged stars as the unbiased estimator of the overall rating of the restaurant.


```python
df_review=pd.read_csv('original_data/yelp_academic_dataset_review.csv')
```

```python
def review_getword(data):
    norm = lambda word: re.sub('[^a-z]', '', word.lower())
    # only include a word once per-review (which de-emphasizes proper nouns)
    reviews = data['text'].values
    wordslist = []
    for review in reviews:
        words = set(norm(word) for word in review.split())
        wordslist.append(words)
    data['words']=wordslist
```

To start the analysis, we do statistics to count how often each word appears in 1, 2, 3, 4 and 5-star reviews.

```python
d_time_star={}
words=df_rest_review['words'].values
stars=df_rest_review['stars'].values
for i in range(len(words)):
    for word in eval(words[i]):
        if word in d_time_star:
            d_time_star[word][0]+=1
            d_time_star[word][int(stars[i])]+=1
        else:
            d_time_star[word]=[0,0,0,0,0,0]
            d_time_star[word][0]+=1
            d_time_star[word][int(stars[i])]+=1
```

```python
for key in d_time_star.keys():
    # we drop the words which appear less than 500 times
    # these words may be misspelled, rare or unreliable
    if d_time_star[key][0]<500:
        d_time_star.pop(key)
s = repr(d_time_star)
f = file('cleaned_data/restaurant_dict_500.txt','w')
f.writelines(s)
f.close()
```

We divide words into 7 groups: the Best (top 5%), the Very-Good (top 10%), the Good (top 20%), the Worst (last 5%), the Very-Bad (last 10%), the Bad (last 20%) and the Neutral. For convenience, we build up a dictionary to look up the assignment of each word.

```python
def good_or_bad(score,cutoffs):
    temp = [i for i in cutoffs]
    temp.append(score)
    temp.sort()
    return temp.index(score)+1
```

```python
def create_great_dict(important_dict):
    values = important_dict.values()
    cutoffs = [np.percentile(values,pct) for pct in [5,10,20,80,90,95]]
    great_dict = {}
    for key in important_dict:
        value = important_dict[key]
        great_dict[key] = {'star':value, 'rate':good_or_bad(value,cutoffs)}
    return great_dict
```

```python
useful_words = df_time_star['word'].values
useful_stars = df_time_star['average-star'].values
important_dict = dict(zip(useful_words,useful_stars))
great_dict = create_great_dict(important_dict)
```

Now we almost transform reviews into quantitative variables, let us see how we get here. Reviews are break down into words, each word is given a score based on its occurance and assigned to certain group, and then each review can be see as a collection of words from different groups.
The explanatory variable (which can be obtained by the function below) is a vector of length 8, the first 7 entries are respectively the number of word occurance in 7 groups, and the last is the total length. The response variable is the number of stars associated with the review. The following part is the selection of classification tool as well as model selection.

```python
def label_count_words(set_of_words):
    rate = []
    for word in set_of_words:
        if word in d_time_star.keys():
            rate.append(great_dict[word]['rate'])
    count = [rate.count(i) for i in [1,2,3,4,5,6,7]]
    total = len(rate)
    # lower the influence of neutral words
    count[3] = 1
    return count,total
```

```python
df_review=pd.read_csv('cleaned_data/restaurantword.csv')
X_raw = []
X_add = []
for i in range(1,700001,700):
    count, total = label_count_words(eval(df_review.loc[i,'words']))
    X_raw.append(count)
    X_add.append(total)
Y = df_review['stars'].values[range(1,700001,700)]
```

```python
X = np.matrix(X_raw)
row_sums = X.sum(axis=1)
X = X.astype(float)/row_sums
X = (X.transpose()).tolist()+[X_add]
X = np.matrix(X).transpose()
```

```python
def split_model(sets):
    features = []
    for i in range(3):
        if i in sets:
            features = features + [i,6-i]
    if 3 in sets:
        features = features + [3]
    if 7 in sets:
        features = features + [7]
    return features
```

```python
X=np.matrix(X)
Y=np.array(Y)
result_dict = {}
for i in range(1,6):
    for sets in combinations([0,1,2,3,7],i):
        features = split_model(sets)
        X_train,X_test,Y_train,Y_test=train_test_split(X[:,features],Y,test_size=0.20,random_state=26)
        predict_dict = {}
        rfclf = RandomForestClassifier(n_estimators=20)
        rfclf.fit(X_train,Y_train)
        predict_dict['RF']=(sum(abs(rfclf.predict(X_test)-Y_test)<=0),sum(abs(rfclf.predict(X_test)-Y_test)<=1))
        dtclf = DecisionTreeClassifier()
        dtclf.fit(X_train,Y_train)
        predict_dict['DT']=(sum(abs(dtclf.predict(X_test)-Y_test)<=0),sum(abs(dtclf.predict(X_test)-Y_test)<=1))
        svmclf = sklearn.svm.SVC(C=10,gamma=1)
        svmclf.fit(X_train,Y_train)
        predict_dict['SVM']=(sum(abs(svmclf.predict(X_test)-Y_test)<=0),sum(abs(svmclf.predict(X_test)-Y_test)<=1))
        lrclf = sklearn.linear_model.LinearRegression()
        lrclf.fit(X_train,Y_train)
        predict_dict['LR']=(sum(abs((lrclf.predict(X_test)+0.5).astype(int)-Y_test)<=0),sum(abs((lrclf.predict(X_test)+0.5).astype(int)-Y_test)<=1))
        
        result_dict[tuple(features)]=predict_dict
result_dict
```

In model fitting trials above, we select here 800 datapoints to train the classifier, and 200 datapoints to test. Since we have sufficient data, we need not to apply cross validation at the moment. Based on the outcomes, we think SVM outperforms other classifiers, especially when the total length of review is excluded. The best model is set manually as [0,1,2,4,5,6]. That means only the top 20% positive/negative are used in prediction. The correctness of prediction is around 50%, and 90% of prediction is close enough.

## VI. Topic Modeling on Reviews

```python
import json
import gensim as gs
from sklearn.feature_extraction.text import CountVectorizer
from gensim.models.ldamodel import LdaModel
```

```python
with open('cleaned_data/docs.json','r') as outfile:
    doc_dict = json.load(outfile)
```

```python
docs = pd.DataFrame(doc_dict.items())
docs.head()
```

```python
vectorizer = CountVectorizer(strip_accents='ascii', stop_words = 'english', min_df=0.0001, max_df=0.1)
X = vectorizer.fit_transform(list(docs[1].values))
corpus = gs.matutils.Sparse2Corpus(X.T)
vocab = {v: k for k, v in vectorizer.vocabulary_.items()}
```

```python
def run_lda(n,n_sims):
    lda = LdaModel(corpus, num_topics=n,alpha='auto', eval_every=1,iterations=n_sims)
    alphas = lda.alpha
    topics = lda.print_topics(n)
    topic_dict = {}
    for t in range(n):
        topic = topics[t]
        portions = topic.split(' +')
        words = [i.split('*')[1] for i in portions]
        words = [str(vocab[int(i)]) for i in words]
        lst = ' '.join(words)
        topic_dict[t]=lst
    return {'lda':lda,'topics':topic_dict,'alpha':alphas}
```

```python
trial = run_lda(n=50,n_sims=10**5)
```

## VII. Conclusion and Discussion

The Yelp dataset contains lots of features for each restaurant. However, most of them are categorical variables, which could potentially be considered bad for making predictions from these variables.

Our approach to tackle the "categorical variables" problem, is to construct continuous variables from the existing dataset. For example, we used the logitude and latitude information to construct a new variable "Distance" to measure the physical distance from the restaurants to city downtown. We thought this distance variable could give more predicting power, but adding this variable does not help much in terms of prediction. Then we tried to construct another new variable "number of competitors" to measure how many restaurants are competing against each other. Again, this new variable does not improve our prediction.

As we experimented the data with all sorts of the models and got insignificant output, we started to wonder: If none of these techniques performs good, what does this tell us? Does it infer the dataset itself is non-informative in terms of predicting stars? The idea struck us! The features we used to fit in the model are all objective/physical descriptions of restaurants, while the restaurants are evaluated very subjectively by users. And this might be the reason why those models performs below our expectation. In other words, if all models are not doing well, then we should question the data.

Luckily, we have the review data for each restaurant. The review data contains information about subjective evaluations from users. We used the review data to predict restaurant stars by adopting test analysis techniques, of which the predicting power outperformed all other features.

Then naturally, we question: what makes the predicting power of reviews so powerful?

In order to answer this question, we use Topic Modelling to analyze the reviews. From the output of Topic Modelling, we can see that the reviews are all about food, not enviroment,  or waiter/waitress, etc. And we do not have the food taste/quality information in the original dataset!
