#### Search for optimal cutting value

```python
def cut(alpha,features = range(11,108)):
    X = data.ix[:,features]
    Y = data.success

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)

    clf = rfc(random_state=0,max_features = 'log2',n_estimators=180,criterion="gini")

    clf.fit(X_train,y_train)
    probs = clf.predict_proba(X_test)[:,1]
    y_pred = probs>alpha
    accuracy = np.mean(y_pred==y_test) 
    return accuracy

scores_cut = []

for i in arange(0.1,1.1,0.1):
    scores_cut.append(cut(i))

    
fig, ax = plt.subplots()
ppl.scatter(ax,arange(0.1,1.1,0.1),scores_cut)

ax.legend()
ax.set_xlabel("Cutting value")
ax.set_ylabel("Score")
ax.set_title("Cutting value v.s. Random Forest Classifier Scores")
```

### 2) Extra Trees Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier as etc
```

```python
def run_etc(n,features = range(9,107),method='gini'):
    X = data.ix[:,range(9,107)]
    Y = pd.factorize(data.stars)[0]
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)

    clf = etc(random_state=0,max_features = 'log2',n_estimators=n,criterion=method,
              min_samples_split=1,oob_score=True,bootstrap=True)

    clf.fit(X_train,y_train,sample_weight=None)
    scores = cross_val_score(clf,X,Y,cv=10)
    return scores

scores_etc_gini = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = run_etc(n=n_estimators[i],features=features_new)
    scores_etc_gini[i,]=scores


scores_etc_entropy = np.zeros((len(n_estimators),10))

for i in range(len(n_estimators)):
    scores = run_etc(n=n_estimators[i],features=features_new,method='entropy')
    scores_etc_entropy[i,]=scores
```

#### Force predicted density to be as expected

One potential trick can be to force predicted relative frequency of stars to be as observed in training data set. For example if we know
that there are roughly 5% of starts = 1, 10% of stars = 2, etc., we can try to produce joint guesses so that they would adhere to these
proportions. This means that some times we will be guessing not according to maximum probability accross categories (if that category
already reached its relative frequency limit, we should not guess it again). It is not clear how to handle this. My idea is that I want
to maximise total sum of probabilities (each category you guess has associated probability with it) with constraint of final joint 
guesses' relative frequencies being equal to the relative frequencies observed in training data set. 

I am implementing this by taking max and second largest probability from each column, then sorting according to difference (difference is
like opportunity cost, because once you run out of category, you will have to guess another one, so if difference between max and
second largest probability is big, you want to get maximum, but if it is not big, you can guess second largest category without loosing
too much on total sum of probabilities).

The reason why this can work, is because we are using additional information. When model makes predictions, it treats each guess 
independently. But we roughly know what proportions are attributed to each star. So when simultaneously predicting many stars, maybe we
can use this knowledge to make more informative decisions.

## V. Reviews and Sentiment Analysis

** Reviews, or opinions of users, eveluate a restaurant from diverse aspects and may provide users with a holistic view besides certain attributes Yelp has included as attributes. In order to explore the information contained in review, we need a different approach to deal with the comparatively large data. The most typical difficulty to deal with reviews is that, they are strings of characters, and not naturally quantitative. Therefore we need to interpret strings by turning them into numbers appropriately. **

It is a natural to start our analysis or strings by breaking down the sentences into words. Since we notice reviews are mostly commentary or sentimental remarks, we expect to see both positive and negative adjectives or adverbs play a leading role. At the same time, we also expect to see some nouns or verbs with certain emotional hints.
<p>
However, those word may be overwhelmed by neutral words whose occurance is much prevalent. We here try to assign a score to each word, and use the score to distinguish meaningful words from meaningless ones. The implementation is to calculate the average rating of reviews in which a certian word appears. All words are ranked from the highest score to the lowest score, and then divided into positive, negative and neutral groups according to their score.
<p>
So far, each review is recognized as a collection of positive words, negative words and neutral words, and is associated with a rating given by the user. We can now apply machine learning toolkits, such as regression or classification on the data, to capture the correlation between review words and the ratings, and then lead to expected inference and prediction. 
<p>
Finally, we notice a piece of review is a random sample from all opinions towards that restaurant, we can expect the review as an unbiased estimator. Indeed, more reviews will lead to better precision in prediction, thus we collect all remarks towards a restaurant and use the averaged stars as the unbiased estimator of the overall rating of the restaurant.


Since we notice that with each review is always a rate, a natural thinking will be that if there is a preference in wording when people describe a good or bad restaurant. For example, if a word appears more often in an 1-star review, it may be reasonable to regard the word as a 'bad description' towards a restaurant. To start the analysis, we do statistics to count how often each word appears in 1, 2, 3, 4 and 5-star reviews.

```python
d_time_star={}
words=df_rest_review['words'].values
stars=df_rest_review['stars'].values
for i in range(len(words)):
    for word in eval(words[i]):
        if word in d_time_star:
            d_time_star[word][0]+=1
            d_time_star[word][stars1[i]]+=1
        else:
            d_time_star[word]=[0,0,0,0,0,0]
            d_time_star[word][0]+=1
            d_time_star[word][stars1[i]]+=1
```

```python
for key in d_time_star.keys():
    # we drop the words which appear less than 500 times
    # these words may be misspelled, rare or unreliable
    if d_time_star[key][0]<500:
        d_time_star.pop(key)
s = repr(d_time_star)
f = file('cleaned_data/restaurant_dict_500.txt','w')
f.writelines(s)
f.close()
```

Here we get a dictionary keyed by words. The most important information contained in this dictionary is the pattern of each words distributed over ratings of different stars. A straightforward and effective way to simplify the pattern is to take an average, and determine the averaged rating of reviews which contain certain word. We here call the averaged rating score of the word.

```python
s = open('cleaned_data/restaurant_dict_500.txt','r').read()
d_time_star = eval(s)
```

```python
df_time_star = pd.DataFrame(d_time_star.items())
df_time_star.columns = ['word', 'count']
```

```python
df_time_star['total'] = [counts[0] for counts in df_time_star['count'].values]
df_time_star['1-star'] = [counts[1] for counts in df_time_star['count'].values]
df_time_star['2-star'] = [counts[2] for counts in df_time_star['count'].values]
df_time_star['3-star'] = [counts[3] for counts in df_time_star['count'].values]
df_time_star['4-star'] = [counts[4] for counts in df_time_star['count'].values]
df_time_star['5-star'] = [counts[5] for counts in df_time_star['count'].values]
df_time_star['1-star-prop'] = df_time_star['1-star']/df_time_star['total']
df_time_star['2-star-prop'] = df_time_star['2-star']/df_time_star['total']
df_time_star['3-star-prop'] = df_time_star['3-star']/df_time_star['total']
df_time_star['4-star-prop'] = df_time_star['4-star']/df_time_star['total']
df_time_star['5-star-prop'] = df_time_star['5-star']/df_time_star['total']
df_time_star['average-star'] = df_time_star['1-star-prop'] + df_time_star['2-star-prop']*2 + df_time_star['3-star-prop']*3 + df_time_star['4-star-prop']*4 + df_time_star['5-star-prop']*5
```

Definitely, we expect to see 'the', 'a', or maybe 'and' in every review, and we always want to get rid of them since we never learn anything from them. We divide words into 7 groups: the Best (top 5%), the Very-Good (top 10%), the Good (top 20%), the Worst (last 5%), the Very-Bad (last 10%), the Bad (last 20%) and the Neutral. For convenience, we build up a dictionary to look up the assignment of each word.

```python
def good_or_bad(score,cutoffs):
    temp = [i for i in cutoffs]
    temp.append(score)
    temp.sort()
    return temp.index(score)+1
```

```python
def create_great_dict(important_dict):
    values = important_dict.values()
    cutoffs = [np.percentile(values,pct) for pct in [5,10,20,80,90,95]]
    great_dict = {}
    for key in important_dict:
        value = important_dict[key]
        great_dict[key] = {'star':value, 'rate':good_or_bad(value,cutoffs)}
    return great_dict
```

```python
useful_words = df_time_star['word'].values
useful_stars = df_time_star['average-star'].values
important_dict = dict(zip(useful_words,useful_stars))
great_dict = create_great_dict(important_dict)
```

Now we almost transform reviews into quantitative variables, let us see how we get here. Reviews are break down into words, each word is given a score based on its occurance and assigned to certain group, and then each review can be see as a collection of words from different groups.
The explanatory variable (which can be obtained by the function below) is a vector of length 8, the first 7 entries are respectively the number of word occurance in 7 groups, and the last is the total length. The response variable is the number of stars associated with the review. The following part is the selection of classification tool as well as model selection.

```python
def label_count_words(set_of_words):
    rate = []
    for word in set_of_words:
        if word in d_time_star.keys():
            rate.append(great_dict[word]['rate'])
    count = [rate.count(i) for i in [1,2,3,4,5,6,7]]
    total = len(rate)
    # lower the influence of neutral words
    count[3] = 1
    return count,total
```

```python
df_review=pd.read_csv('cleaned_data/restaurantword.csv')
X_raw = []
X_add = []
for i in range(1,700001,700):
    count, total = label_count_words(eval(df_review.loc[i,'words']))
    X_raw.append(count)
    X_add.append(total)
Y = df_review['stars'].values[range(1,700001,700)]
```

```python
X = np.matrix(X_raw)
row_sums = X.sum(axis=1)
X = X.astype(float)/row_sums
X = (X.transpose()).tolist()+[X_add]
X = np.matrix(X).transpose()
```

```python
def split_model(sets):
    features = []
    for i in range(3):
        if i in sets:
            features = features + [i,6-i]
    if 3 in sets:
        features = features + [3]
    if 7 in sets:
        features = features + [7]
    return features
```

```python
X=np.matrix(X)
Y=np.array(Y)
result_dict = {}
for i in range(1,6):
    for sets in combinations([0,1,2,3,7],i):
        features = split_model(sets)
        X_train,X_test,Y_train,Y_test=train_test_split(X[:,features],Y,test_size=0.20,random_state=26)
        predict_dict = {}
        rfclf = RandomForestClassifier(n_estimators=20)
        rfclf.fit(X_train,Y_train)
        predict_dict['RF']=(sum(abs(rfclf.predict(X_test)-Y_test)<=0),sum(abs(rfclf.predict(X_test)-Y_test)<=1))
        dtclf = DecisionTreeClassifier()
        dtclf.fit(X_train,Y_train)
        predict_dict['DT']=(sum(abs(dtclf.predict(X_test)-Y_test)<=0),sum(abs(dtclf.predict(X_test)-Y_test)<=1))
        svmclf = sklearn.svm.SVC(C=10,gamma=1)
        svmclf.fit(X_train,Y_train)
        predict_dict['SVM']=(sum(abs(svmclf.predict(X_test)-Y_test)<=0),sum(abs(svmclf.predict(X_test)-Y_test)<=1))
        lrclf = sklearn.linear_model.LinearRegression()
        lrclf.fit(X_train,Y_train)
        predict_dict['LR']=(sum(abs((lrclf.predict(X_test)+0.5).astype(int)-Y_test)<=0),sum(abs((lrclf.predict(X_test)+0.5).astype(int)-Y_test)<=1))
        
        result_dict[tuple(features)]=predict_dict
result_dict
```

In model fitting trials above, we select here 800 datapoints to train the classifier, and 200 datapoints to test. Since we have sufficient data, we need not to apply cross validation at the moment. Based on the outcomes, we think SVM outperforms other classifiers, especially when the total length of review is excluded. The best model is set manually as [0,1,2,4,5,6]. That means only the top 20% positive/negative are used in prediction. The correctness of prediction is around 50%, and 90% of prediction is close enough.

## VI. Topic Modeling on Reviews

```python
import json
import gensim as gs
from sklearn.feature_extraction.text import CountVectorizer
from gensim.models.ldamodel import LdaModel
```

```python
with open('cleaned_data/docs.json','r') as outfile:
    doc_dict = json.load(outfile)
```

```python
docs = pd.DataFrame(doc_dict.items())
docs.head()
```

```python
vectorizer = CountVectorizer(strip_accents='ascii', stop_words = 'english', min_df=0.0001, max_df=0.1)
X = vectorizer.fit_transform(list(docs[1].values))
corpus = gs.matutils.Sparse2Corpus(X.T)
vocab = {v: k for k, v in vectorizer.vocabulary_.items()}
```

```python
def run_lda(n,n_sims):
    lda = LdaModel(corpus, num_topics=n,alpha='auto', eval_every=1,iterations=n_sims)
    alphas = lda.alpha
    topics = lda.print_topics(n)
    topic_dict = {}
    for t in range(n):
        topic = topics[t]
        portions = topic.split(' +')
        words = [i.split('*')[1] for i in portions]
        words = [str(vocab[int(i)]) for i in words]
        lst = ' '.join(words)
        topic_dict[t]=lst
    return {'lda':lda,'topics':topic_dict,'alpha':alphas}
```

## VII. Conclusion and Discussion

The Yelp dataset contains lots of features for each restaurant. However, most of them are categorical variables, which could potentially be considered bad for making predictions from these variables.

Our approach to tackle the "categorical variables" problem, is to construct continuous variables from the existing dataset. For example, we used the logitude and latitude information to construct a new variable "Distance" to measure the physical distance from the restaurants to city downtown. We thought this distance variable could give more predicting power, but adding this variable does not help much in terms of prediction. Then we tried to construct another new variable "number of competitors" to measure how many restaurants are competing against each other. Again, this new variable does not improve our prediction.

As we experimented the data with all sorts of the models and got insignificant output, we started to wonder: If none of these techniques performs good, what does this tell us? Does it infer the dataset itself is non-informative in terms of predicting stars? The idea struck us! The features we used to fit in the model are all objective/physical descriptions of restaurants, while the restaurants are evaluated very subjectively by users. And this might be the reason why those models performs below our expectation. In other words, if all models are not doing well, then we should question the data.

Luckily, we have the review data for each restaurant. The review data contains information about subjective evaluations from users. We used the review data to predict restaurant stars by adopting test analysis techniques, of which the predicting power outperformed all other features.

Then naturally, we question: what makes the predicting power of reviews so powerful?

In order to answer this question, we use Topic Modelling to analyze the reviews. From the output of Topic Modelling, we can see that the reviews are all about food, not enviroment,  or waiter/waitress, etc. And we do not have the food taste/quality information in the original dataset!
